```python
import pandas as pd
from google.colab import drive 
drive = drive.mount('/content/drive') 
```

```python
project = 'elasticsearch'

df = pd.read_csv(f"drive/MyDrive/subtract/{project}/class.csv")
```

```python
df
```

```python
df.drop('Hash', axis=1, inplace=True)
df.drop('LongName', axis=1, inplace=True)
```

```python
df
```

```python
old_df = df.columns
for col in old_df:
  if df[col].max() == df[col].min():
    df.drop(col, axis=1, inplace=True)
```

```python
df
```

```python
from sklearn.model_selection import StratifiedShuffleSplit
 
split = StratifiedShuffleSplit(n_splits=1, test_size=0.2)

nb = df['Number of Bugs']
 
for train_index, test_index in split.split(df, nb):
  train_set = df.loc[train_index]
  test_set = df.loc[test_index]
```

```python
train_set
```

```python
test_set
```

```python
df['Number of Bugs'].value_counts().plot(kind='bar')
```

```python
train_set['Number of Bugs'].value_counts().plot(kind='bar')
```

```python
test_set['Number of Bugs'].value_counts().plot(kind='bar')
```

```python
def normalizeValueOfTarget(value):
  if value == 0 :
    return 0
  if value == 1:
    return 1
  else: 
     return 2
     
train_set['Number of Bugs'] = train_set['Number of Bugs'].apply(normalizeValueOfTarget)
test_set['Number of Bugs'] = test_set['Number of Bugs'].apply(normalizeValueOfTarget)
```

```python
train_set['Number of Bugs'].value_counts().plot(kind='bar')
```

```python
from sklearn.utils import resample

maxcount = 6000

train_resampled = train_set[0:0]

for grp in train_set['Number of Bugs'].unique():
    GrpDF = train_set[train_set['Number of Bugs'] == grp]
    resampled = resample(GrpDF, replace=True, n_samples=int(maxcount), random_state=700)
    train_resampled = train_resampled.append(resampled)
```

```python
train_resampled['Number of Bugs'].value_counts().plot(kind='bar')
```

```python
test_set['Number of Bugs'].value_counts().plot(kind='bar')
```

```python
from sklearn.utils import resample

maxcount = 1500

test_resampled = test_set[0:0]

for grp in test_set['Number of Bugs'].unique():
    GrpDF = test_set[test_set['Number of Bugs'] == grp]
    resampled = resample(GrpDF, replace=True, n_samples=int(maxcount), random_state=700)
    test_resampled = test_resampled.append(resampled)
```

```python
test_resampled['Number of Bugs'].value_counts().plot(kind='bar')
```

```python
columnsToNormalize = list(train_resampled.columns.values)
for column in columnsToNormalize:
  minValueOfCol = train_resampled[column].min()
  maxValueOfCol = train_resampled[column].max()
  
  train_resampled[column] = (train_resampled[column] - minValueOfCol) / (maxValueOfCol-minValueOfCol)
  test_resampled[column] = (test_resampled[column] - minValueOfCol) / (maxValueOfCol-minValueOfCol)
```

```python
def normalizeValueOfTarget(value):
  if value == 0 :
    return 0
  if value == 1:
    return 1
  else: 
     return 2
     
train_resampled_NB = train_resampled_NB.apply(normalizeValueOfTarget)
test_resampled_NB = test_resampled_NB.apply(normalizeValueOfTarget)
```
